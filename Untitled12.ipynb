```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('/content/Unemployment in India.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['longitude'] = np.nan
data['latitude'] = np.nan
```

```python
data.columns = ['States', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate', 'Region', 'longitude', 'latitude']
```

```python
data.head()
```

```python
numeric_data = data.select_dtypes(include=[np.number])

```

```python
if numeric_data.shape[1] == 0:
    print("No numeric data available to plot heatmap.")
```

```python
plt.figure(figsize=(8, 6))
ax = sns.heatmap(numeric_data.corr(), annot=True, linewidths=3, cmap='coolwarm')
ax.tick_params(size=10, color='w', labelsize=10, labelcolor='w')
plt.show()
```

```python
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed', hue='Region', data=data)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title("Indian Unemployment")
sns.histplot(x='Estimated Unemployment Rate', hue='Region', data=data)
plt.show
```

```python
unemployment = data[['States', 'Region', 'Estimated Unemployment Rate']].dropna(subset=['States', 'Region', 'Estimated Unemployment Rate'])
```

```python
if unemployment.empty:
    print("No valid data available for plotting.")
```

```python
figure = px.sunburst(
    unemployment,
    path=['Region', 'States'],
    values='Estimated Unemployment Rate',
    width=800,
    height=800,
    color='Estimated Unemployment Rate',
    color_continuous_scale='RdYlGn',
    title="Unemployment Rate in India"
)
figure.show()
```
